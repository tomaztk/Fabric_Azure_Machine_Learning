# Module 1 

Microsoft Fabric is an all-purpose (sass) analytics platform designed to manage end-to-end data workflows, including data integration, data engineering, real-time analytics, business intelligence (BI), data science, and AI. It brings together multiple capabilities such as Azure Synapse, Power BI, Data Lake, Data Factory, and Machine Learning under a unified architecture.

To get started with Microsoft Fabric, it is essential to become familiar with the following general concepts:

## Main Concepts

### Lakehouse:

A unified storage layer that combines data lakes and data warehouses, providing a single source of truth for data storage.

Fabric uses Delta Lake for transactional support, providing ACID transactions and scalable data storage.

### Data Pipeline:

The process of automating the movement and transformation of data from raw sources to consumable insights.

In Microsoft Fabric, Pipelines are used to orchestrate data flows between different storage layers, including Data Lake, Lakehouse, and Power BI datasets.

### Workspaces:

Workspaces are organizational units within Fabric that manage resources and assets such as notebooks, datasets, and pipelines.

### Dataflow:

A component of Power BI, a Dataflow allows users to define and reuse ETL (Extract, Transform, Load) logic that transforms data from various sources.

### Notebooks:

Notebooks in Microsoft Fabric allow users to write code (in Python, PySpark, or SQL) for data processing, transformation, and machine learning tasks.

### MLFlow Integration:

MLFlow helps with the management of machine learning models, including model tracking, versioning, and deployment to different environments like test, dev, and prod.
